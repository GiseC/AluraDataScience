#1. Análisis de facturación



```python
#funcion para encontrar la facturacion de cada tienda
def calcular_facturacion(tienda):
    tienda['Facturación'] = tienda['Precio'] + tienda['Costo de envío']
    return tienda

#se crea la columna para cada tienda
tiendaA = calcular_facturacion(tienda)
tiendaB = calcular_facturacion(tienda2)
tiendaC = calcular_facturacion(tienda3)
tiendaD = calcular_facturacion(tienda4)

#funcion para analizar valores de cada tienda, total, maximo, minimo y promedio de facturacion
def total(tiendax):
    lista_valores = []
    totalFacturacion = tiendax['Facturación'].sum()
    maximo = tiendax['Facturación'].max()
    minimo = tiendax['Facturación'].min()
    promedio = round(tiendax['Facturación'].mean(),2)
    #se agregan los 4 valores a una lista
    lista_valores.append(totalFacturacion)
    lista_valores.append(maximo)
    lista_valores.append(minimo)
    lista_valores.append(promedio)
    #se devuelve una lista con los 4 valores de una tienda
    return lista_valores

#se agregan cada lista de tienda con sus valores como item de otra lista
#totalventas es una lista con 4 listas dentro, una por cada tienda conteniendo los valores analizados
totalventas = []
totalventas.append(total(tiendaA))
totalventas.append(total(tiendaB))
totalventas.append(total(tiendaC))
totalventas.append(total(tiendaD))

#se recorre totalventas mostrando los valores de cada tienda
for i in range(len(totalventas)):
    k = i+1
    print(f"Facturación tienda {k}: es {totalventas[i][0]}")
    print(f"Maxima venta de la tienda {k}: es {totalventas[i][1]}")
    print(f"Minima venta de la tienda {k}: es {totalventas[i][2]}")
    print(f"Promedio de venta de la tienda {k}: es {totalventas[i][3]}")
    print("************************************")

#funcion para encontrar el menor, mayor y media de todas las tiendas
def mayorvalor(lista):
    mayor = lista[0][0]
    menor= lista[0][0]
    indice = 0
    indicem = 0
    for i in range(len(lista)):
        if lista[i][0] > mayor:
            mayor = lista[i][0]
            indice = i
        elif lista[i][0] < menor:
            menor = lista[i][0]
            indicem = i
    print(f"La tienda con mayor facturación es la tienda {indice+1} con un total de {mayor}")
    print(f"La tienda con menor facturación es la tienda {indicem+1} con un total de {menor}")
    promedio = sum([lista[i][0] for i in range(len(lista))])/len(lista)
    print("El promedio general de ventas es: ", promedio)

#llamo a la ultima funcion donde me muestra el resultado general para las tiendas
mayor = mayorvalor(totalventas)
```

# 2. Ventas por categoría

```python
#funcion para agrupar total de facturacion por categoria por tienda
def venta_cat(tiendax):
  return tiendax.groupby('Categoría del Producto')['Facturación'].sum()

#funcion para agrupar cantidad de cada categoria por tienda
def cantidad_cat(tiendax):
    return tiendax.groupby('Categoría del Producto')['Categoría del Producto'].count()

#creo lista con los valores aportados por la funcion
tiendas = [tienda, tienda2, tienda3, tienda4]  # Lista de tiendas
listas_tiendas = []  # Lista para almacenar las listas de cada tienda

#creo lista de tuplas para almacenar categoria, total de ventas y cantidad
#recorro cada tienda y aplico las funciones para obtener las listas
for tiendax in tiendas:
    ventas_categorias = venta_cat(tiendax)
    cantidad_categorias = cantidad_cat(tiendax)
    #creo la lista de tuplas con los valores obtenidos, utilizo zip
    lista_tienda = [(categoria, total_venta, cantidad) for (categoria, total_venta), (categoria_cantidad, cantidad) in zip(ventas_categorias.items(), cantidad_categorias.items()) if categoria == categoria_cantidad]
    listas_tiendas.append(lista_tienda)

listaA, listaB, listaC, listaD = listas_tiendas  # Asignar las listas a variables separadas

#muestro resultados para cada tienda
k = 1
for i in listas_tiendas:
  print("******Tienda ",k,"**********")
  k += 1
  for j in i:
    print(j[0], "  total: ",j[1], "  Cantidad: ", j[2])

```

```python
#mostrando lo anterior en formato tabla

#funcion para agrupar por producto por tienda
def cantidad_prod(tiendax):
    return tiendax.groupby('Producto')['Producto'].count()

#creo lista con los valores aportados por la funcion
tiendas = [tienda, tienda2, tienda3, tienda4]  # Lista de tiendas
prod_tiendas_dict = {} # Diccionario para almacenar los resultados por tienda

for i, tiendax in enumerate(tiendas):
    cantidad_productos = cantidad_prod(tiendax)
    # Convertir la serie a un diccionario para facilitar la creación del DataFrame
    prod_tiendas_dict[f'Tienda {i+1}'] = cantidad_productos.to_dict()

# Crear un DataFrame a partir del diccionario
df_productos_por_tienda = pd.DataFrame(prod_tiendas_dict)

# Mostrar la tabla
display(df_productos_por_tienda)
```

```python
import matplotlib.pyplot as plt

# basandome en el codigo anterior creo el grafico de lineas

# se crea un diccionario para guardar los productos
producto_cantidad = {}
for i, tienda_lista in enumerate(prod_tiendas):
    for producto, cant in tienda_lista:
        if producto not in producto_cantidad:
            producto_cantidad[producto] = {}
        producto_cantidad[producto][f'Tienda {i+1}'] = cant

# se convierte el diccionario a un dataframe
df_products = pd.DataFrame(producto_cantidad).T.fillna(0)

# Calcular la cantidad total vendida por producto y ordenar el DataFrame
df_products['Total Vendido'] = df_products.sum(axis=1)
df_products_sorted = df_products.sort_values(by='Total Vendido', ascending=False) # Ordenar de mayor a menor

colores = ['#023047', '#FFB703', '#219EBC', '#FB8500']
tiendas_nombres = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# cargo el eje x con los nombres de productos ordenados
productos_ordenados = df_products_sorted.index

# personalizo grafico
plt.figure(figsize=(12, 7))

for i, tiendax in enumerate(tiendas_nombres): # Iterar sobre los nombres de las tiendas
    colorL = colores[i]
    tiendaN = tiendas_nombres[i]
    plt.plot(productos_ordenados, df_products_sorted[tiendax], marker='o', linestyle='-', color=colorL, label=tiendaN)

# Agreo etiquetas y titulos
plt.xlabel('Producto', fontweight='bold')
plt.ylabel('Cantidad Vendida', fontweight='bold')
plt.title('Cantidad de Productos Vendidos por Tienda (Gráfico de Líneas - Ordenado por Total Vendido)', fontweight='bold')

# cambio eje x
plt.xticks(rotation=45, ha='right')
plt.grid(True)

# agrego leyenda
plt.legend()
plt.tight_layout()
plt.show()
```
